```python
#!pip install beautifulsoup4
from bs4 import BeautifulSoup as bs

def get_players_tournament(response, nome_torneo, year,numero):

    lista=[]
    soup = bs(response.text)

    list = soup.find('select', id ="matchPlayer-filter")

    players_tournament = list.findAll('option')

    #data = soup.find("div", class_ ="date-location").findAll("span")[1]#.get_text(strip=True)
    #print(data)

    for p in players_tournament[1:]:
        player_name = p.get_text(strip=True)
        player_id = p['value']
        lista.append([year, nome_torneo, numero, player_name, player_id])

    return lista
```
