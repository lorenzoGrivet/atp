```python
#!pip install beautifulsoup4
from bs4 import BeautifulSoup as bs
soup = bs(response.text)
```

```python
list = soup.findAll('div',class_="atp_accordion-item")
```

```python
for l in list:
    incontri = l.findAll('div', class_="match-stats")
    round_name = l.find("div", class_="tournament-day").find("h4").get_text(strip=True)
    print(f"\n{round_name}")

    for match in incontri:
        names = match.find_all("div", class_="name")

        w_a = names[0].find("a")
        l_a = names[1].find("a")

        winner_name = w_a.text
        loser_name  = l_a.text

        winner_href = w_a.get("href")
        loser_href  = l_a.get("href")

        winner_code = winner_href.split("/")[4]
        loser_code  = loser_href.split("/")[4]

        print(winner_name, winner_code, "vs", loser_name, loser_code)
```
